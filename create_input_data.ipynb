#완성본

```python
def input_preprocessing(input_path, output_path):
  data = pd.read_csv(input_path, usecols = [0, 4, 5, 6], header = None
                            , names = ['timestamp', 'Roll', 'Pitch', 'Heading'])
  data = data[1:]

# dt 구하기
  time = pd.read_csv(input_path, usecols = [0], header = None)
  time = time[1:]
  time = pd.to_datetime(time[0].astype(str))
  time = time.diff()
  time_diff = time.dt.total_seconds().astype(str).str.split(' ').str[-1]

# 위치 x 변화량 구하기
  pos_x = pd.read_csv(input_path, usecols = [1], header = None)
  pos_x = pos_x[1:]
  pos_x =pos_x.astype(float)
  data['dX'] = pos_x.diff()

# 위치 y 변화량 구하기
  pos_y = pd.read_csv(input_path, usecols = [2], header = None)
  pos_y = pos_y[1:]
  pos_y =pos_y.astype(float)
  data['dY'] = pos_y.diff()

# 속도 변화량 구하기
  velocity = pd.DataFrame()
  velocity['dt'] = time_diff.astype(float)
  velocity['dX'] = data['dX']
  velocity['vel_X'] = velocity['dX'] / velocity['dt']
  velocity['dv_X'] = velocity['vel_X'].diff()
  data['dv_X'] = velocity['dv_X']

  velocity['dY'] = data['dY']
  velocity['vel_Y'] = velocity['dY'] / velocity['dt']
  velocity['dv_Y'] = velocity['vel_Y'].diff()
  data['dv_Y'] = velocity['dv_Y']
  data = data[2:]

  data.to_csv(output_path, index=False)
```

```python
input_preprocessing('/content/120219_new_odo_mu_100.csv', '/content/120219_odo.csv')
input_preprocessing('/content/120317_new_odo_mu_100.csv', '/content/120317_odo.csv')
```
